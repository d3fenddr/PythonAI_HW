```python
import numpy as np
```

## Task 1: 4×4 Array Operations


```python
np.random.seed(42)
array_4x4 = np.random.randint(1, 51, size=(4, 4))
print("4×4 Array:")
print(array_4x4)
```

```python
row_sums = np.sum(array_4x4, axis=1)
column_sums = np.sum(array_4x4, axis=0)
overall_mean = np.mean(array_4x4)

print(f"Row sums: {row_sums}")
print(f"Column sums: {column_sums}")
print(f"Overall mean: {overall_mean:.2f}")
```

## Task 2: Array Transformation


```python
array_3x5 = np.arange(15).reshape(3, 5)
print("Original 3×5 Array:")
print(array_3x5)
print()
```

```python
array_transposed = array_3x5.T
print("Transposed Array (5×3):")
print(array_transposed)
print()
```

```python
array_5x3 = array_3x5.reshape(5, 3)
print("Reshaped Array (5×3):")
print(array_5x3)
```

## Task 3: Sales Report

In this section, we compute totals for simple data.


```python
sales = [120, 135, 100, 150, 90, 200, 170]

total_sales = sum(sales)
average_sales = sum(sales) / len(sales)
maximum_sales = max(sales)
minimum_sales = min(sales)

print(f"Sales data: {sales}")
print(f"\nTotal sales for the week: {total_sales}")
print(f"Average value: {average_sales:.2f}")
print(f"Maximum value: {maximum_sales}")
print(f"Minimum value: {minimum_sales}")
```
