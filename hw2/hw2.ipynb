## Задание 1: Данные о книгах


```python
books = pd.DataFrame({
    'Название': ['Война и мир', 'Преступление и наказание', 'Мастер и Маргарита', 
                 'Евгений Онегин', '1984', 'Анна Каренина'],
    'Автор': ['Толстой Л.Н.', 'Достоевский Ф.М.', 'Булгаков М.А.', 
              'Пушкин А.С.', 'Оруэлл Дж.', 'Толстой Л.Н.'],
    'Год издания': [2010, 2016, 2018, 2012, 2019, 2020],
    'Цена': [850, 620, 490, 380, 550, 720]
})

print(books)
```

```python
avg_price = books['Цена'].mean()
print(f"Средняя цена книг: {avg_price:.2f}")
```

```python
books_after_2015 = books[books['Год издания'] > 2015]
print("Книги, изданные после 2015 года:")
print(books_after_2015)
```

```python
books_sorted = books.sort_values('Цена')
print("Книги, отсортированные по цене:")
print(books_sorted)
```

## Задание 2: Данные о заказах из CSV


```python
orders = pd.read_csv('orders.csv')
print("Первые 10 строк:")
print(orders.head(10))
```

```python
orders_per_client = orders['Клиент'].value_counts()
print("Количество заказов каждого клиента:")
print(orders_per_client)
```

```python
max_sum = orders['Сумма'].max()
min_sum = orders['Сумма'].min()
print(f"Максимальная сумма заказа: {max_sum}")
print(f"Минимальная сумма заказа: {min_sum}")
```

```python
total_sum = orders['Сумма'].sum()
print(f"Общая сумма всех заказов: {total_sum}")
```

## Задание 3: Данные о продуктах питания


```python
food = pd.DataFrame({
    'Продукт': ['Яблоко', 'Банан', 'Курица', 'Говядина', 'Рис', 'Гречка', 
                'Молоко', 'Йогурт', 'Хлеб', 'Макароны'],
    'Категория': ['Фрукты', 'Фрукты', 'Мясо', 'Мясо', 'Крупы', 'Крупы', 
                  'Молочные', 'Молочные', 'Хлебобулочные', 'Крупы'],
    'Калории': [52, 89, 335, 250, 130, 343, 64, 59, 265, 371],
    'Белки': [0.3, 1.1, 31, 26, 2.7, 12.6, 3.2, 4.3, 9, 13]
})

print(food)
```

```python
high_cal = food[food['Калории'] > 300]
print("Продукты с калорийностью выше 300:")
print(high_cal)
```

```python
avg_protein = food.groupby('Категория')['Белки'].mean()
print("Среднее количество белков по категориям:")
print(avg_protein)
```

```python
food_sorted = food.sort_values('Калории', ascending=False)
print("Продукты, отсортированные по калорийности (по убыванию):")
print(food_sorted)
```

## Задание 4: Данные о сотрудниках и проектах


```python
employees = pd.DataFrame({
    'Имя': ['Иванов', 'Петров', 'Сидоров', 'Иванов', 'Козлов', 'Петров', 'Смирнов', 'Сидоров'],
    'Проект': ['Alpha', 'Beta', 'Alpha', 'Gamma', 'Beta', 'Alpha', 'Gamma', 'Beta'],
    'Часы': [40, 35, 25, 30, 45, 28, 50, 32]
})

print(employees)
```

```python
hours_per_employee = employees.groupby('Имя')['Часы'].sum()
print("Общее количество часов по каждому сотруднику:")
print(hours_per_employee)
```

```python
hours_per_project = employees.groupby('Проект')['Часы'].sum()
print("Общее количество часов по каждому проекту:")
print(hours_per_project)
```

```python
top_employee = hours_per_employee.idxmax()
max_hours = hours_per_employee.max()
print(f"Сотрудник с наибольшим количеством часов: {top_employee} ({max_hours} часов)")
```

## Задание 5: Продажи билетов в кино


```python
tickets = pd.DataFrame({
    'Фильм': ['Дюна 2', 'Дюна 2', 'Оппенгеймер', 'Оппенгеймер', 'Барби', 'Барби', 
              'Дюна 2', 'Оппенгеймер', 'Барби', 'Дюна 2', 'Оппенгеймер', 'Барби'],
    'Город': ['Москва', 'Санкт-Петербург', 'Москва', 'Казань', 'Москва', 'Санкт-Петербург',
              'Казань', 'Санкт-Петербург', 'Казань', 'Новосибирск', 'Новосибирск', 'Новосибирск'],
    'Проданные билеты': [450, 320, 280, 150, 380, 290, 200, 240, 160, 180, 120, 210]
})

print(tickets)
```

```python
tickets_per_film = tickets.groupby('Фильм')['Проданные билеты'].sum()
print("Общее количество проданных билетов по каждому фильму:")
print(tickets_per_film)
```

```python
tickets_per_city = tickets.groupby('Город')['Проданные билеты'].sum()
print("Общее количество билетов по каждому городу:")
print(tickets_per_city)
```

```python
top_film = tickets_per_film.idxmax()
max_tickets = tickets_per_film.max()
print(f"Фильм с наибольшим количеством продаж: {top_film} ({max_tickets} билетов)")
```
